```python
def plot_diheds(data, N=50, title=None, ax1=None, ax2=None, color=None):
    if ax1 is None or ax2 is None:
        gs = gridspec.GridSpec(2, 6)
        ax1 = plt.subplot(gs[:1, :2], polar=True)
        ax2 = plt.subplot(gs[:1, 2:])

    x  = data
    
    sns.distplot(x, bins=N, ax=ax2, color=color, kde=True)
    radii, theta = np.histogram(x, bins=N, normed=True)
    ax1.set_yticklabels([])
    ax1.set_xticklabels([0, 45, 90, 135, 180, -135, -90, -45])
    
    ax2.set_yticks([])
    ax2.set(xlabel='Angle', ylabel='Density')
    cleanup_top_right_axes(ax2)
    _=plt.xlim([-180, 180])
    width = (2*np.pi) / N

    bars = ax1.bar(np.pi * theta[1:] / 180., radii, width=width, color=color, alpha=.5)

    if title is not None:
        plt.suptitle(title)

    plt.tight_layout()
    
    f = plt.gcf()
    return f, (ax1, ax2)
```

```python
plot_diheds(np.rad2deg(diheds[:, 1]));
```

```python
meta, dtrajs = load_trajs('dtrajs')
```

```python
dtrajs.get(0)
```

```python
for t in meta['type'].unique():
    print(t)
```

```python
data_dict = {}  # This will have three keys, one for each sim type
for t in meta['type'].unique():
    indexes = meta[meta['type'] == t].index  # The simulation indexes
    dihed_arr = np.concatenate([dtrajs.get(key) for key in indexes])
    data_dict[t] = dihed_arr
```

```python
wtapoup_ctni = np.rad2deg(data_dict['WT apo uP'][:, 0])
wtapoup_ctnt = np.rad2deg(data_dict['WT apo uP'][:, 1])
```

```python
wtapop_ctni = np.rad2deg(data_dict['WT apo P'][:, 0])
wtapop_ctnt = np.rad2deg(data_dict['WT apo P'][:, 1])
```

```python
sns.distplot(wtapop_ctnt, kde=False)
```

```python
from plot_utils import figure_dims
```

```python
import matplotlib.patches as mpatches
```

```python
uP = mpatches.Patch(label='WT apo uP')
P = mpatches.Patch(color='tab:orange', label='WT apo P')
```

```python
f, (ax1, ax2) = plot_diheds(wtapoup_ctnt, color='tab:blue')
plot_diheds(wtapop_ctnt, ax1=ax1, ax2=ax2, color='tab:orange')
f.suptitle('cTnT dihedral angle')
plt.legend(handles =[uP, P], loc='best')
f.set_size_inches(figure_dims(600, 0.9))
f.tight_layout()
f.savefig('cTnT-dihed-angle.pdf')
```

```python
f, (ax1, ax2) = plot_diheds(wtapoup_ctni, color='tab:blue')
plot_diheds(wtapop_ctni, ax1=ax1, ax2=ax2, color='tab:orange')
f.suptitle('cTnI dihedral angle')
plt.legend(handles =[uP, P], loc='best')
f.set_size_inches(figure_dims(600, 0.9))
f.tight_layout()
f.savefig('cTnI-dihed-angle.pdf')
```
